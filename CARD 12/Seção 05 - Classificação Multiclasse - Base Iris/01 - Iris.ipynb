```python
import pandas as pd
import numpy as np
import tensorflow as tf
import sklearn
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import utils as np_utils
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
base = pd.read_csv('iris.csv')
base # 150 x 5
```

```python
x = base.iloc[:, 0:4].values # : = todas as linhas # colunas 0 a 4 (3) menos o nome
print(x)
```

```python
y = base.iloc[:, 4].values # todas as linhas, só a coluna do nome
print(y)
```

```python
label_encoder = LabelEncoder() # transforma as strings em numeros
y = label_encoder.fit_transform(y)
print(y)
```

```python
y = np_utils.to_categorical(y) # transforma os numeros em categorias
y
```

```python
x_treinamento, x_teste, y_treinamento, y_teste = train_test_split(x, y, test_size = 0.25)
```

```python
x_treinamento.shape, x_teste.shape # 112 para treinar e 38 para testar
```

```python
y_treinamento.shape, y_teste.shape # 112 para treinar e 38 para testar
```

## rede neural

```python
rede_neural = Sequential([
    tf.keras.layers.InputLayer(shape=(4,)),
    tf.keras.layers.Dense(units = 4, activation='relu'),
    tf.keras.layers.Dense(units = 4, activation='relu'),
    tf.keras.layers.Dense(units = 3, activation='softmax'), # 3 por causa das 3 classes, softmax pela mesma causa
])
rede_neural.summary()
```

```python
rede_neural.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
```

```python
rede_neural.fit(x_treinamento, y_treinamento, batch_size=10, epochs = 1000)
```

```python
rede_neural.evaluate(x_teste, y_teste) # [0] valor da perda, [1] precisão
```

```python
previsoes = rede_neural.predict(x_teste)
previsoes
```

```python
previsoes = previsoes > 0.5
print(previsoes)
```

```python
y_teste2 = [np.argmax(t) for t in y_teste]
print(y_teste2)
```

```python
previsoes2 = [np.argmax(t) for t in previsoes]
print(previsoes2)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_teste2, previsoes2)
```

```python
confusion_matrix(y_teste2, previsoes2)
```
