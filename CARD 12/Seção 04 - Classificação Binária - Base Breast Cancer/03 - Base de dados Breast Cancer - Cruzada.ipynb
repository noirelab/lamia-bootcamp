```python
import pandas as pd
import tensorflow as tf
import sklearn
import scikeras
```

```python
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score
from tensorflow.keras.models import Sequential
from tensorflow.keras import backend as k
```

```python
x = pd.read_csv('entradas_breast.csv')
y = pd.read_csv('saidas_breast.csv')
```

```python
def criar_rede():
    k.clear_session() # backend do tf, tem que limpar a sessão antes de criar a estrutura da rede neural
    rede_neural = Sequential([
        tf.keras.layers.InputLayer(input_shape=(30,)), # entrada com 30 neuronios
        tf.keras.layers.Dense(units=16, activation='relu', kernel_initializer='random_uniform'), # 1 oculta
        tf.keras.layers.Dropout(rate = 0.2), # 20% de dropout
        tf.keras.layers.Dense(units=16, activation='relu', kernel_initializer='random_uniform'), # 2 oculta
        tf.keras.layers.Dropout(rate = 0.2), # 20% de dropout
        tf.keras.layers.Dense(units=1, activation='sigmoid') # saida
    ])
    otimizador = tf.keras.optimizers.Adam(learning_rate=0.001, clipvalue=0.5) # nenhum peso pode ser maior que o clipvalue
    rede_neural.compile(optimizer=otimizador, loss='binary_crossentropy', metrics=['binary_accuracy'])

    return rede_neural
```

```python
rede_neural = KerasClassifier(model = criar_rede, epochs = 100, batch_size = 10)
```

```python
resultados = cross_val_score(estimator = rede_neural, X = x, y = y, cv = 10, scoring = 'accuracy')
```

```python
# com dropout
print(resultados)
print(resultados.mean())
print(resultados.std())
```
