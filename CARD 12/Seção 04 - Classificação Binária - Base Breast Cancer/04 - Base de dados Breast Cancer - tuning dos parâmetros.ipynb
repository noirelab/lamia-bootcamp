```python
import pandas as pd
import tensorflow as tf
import sklearn
import scikeras
```

```python
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score
from tensorflow.keras import backend as k
from tensorflow.keras.models import Sequential
from sklearn.model_selection import GridSearchCV
```

```python
x = pd.read_csv('entradas_breast.csv')
y = pd.read_csv('saidas_breast.csv')
```

```python
def criar_rede(optimizer, loss, kernel_initializer, activation, neurons):
    k.clear_session() # backend do tf, tem que limpar a sessão antes de criar a estrutura da rede neural
    rede_neural = Sequential([
        tf.keras.layers.InputLayer(input_shape=(30,)), # entrada com 30 neuronios
        tf.keras.layers.Dense(units=neurons, activation=activation, kernel_initializer=kernel_initializer), # 1 oculta
        tf.keras.layers.Dropout(rate = 0.2), # 20% de dropout
        tf.keras.layers.Dense(units=neurons, activation=activation, kernel_initializer=kernel_initializer), # 2 oculta
        tf.keras.layers.Dropout(rate = 0.2), # 20% de dropout
        tf.keras.layers.Dense(units=1, activation='sigmoid') # saida
    ])
    rede_neural.compile(optimizer=optimizer, loss=loss, metrics=['binary_accuracy'])

    return rede_neural
```

```python
rede_neural = KerasClassifier(model = criar_rede)
```

```python
parametros = {
    'batch_size': [10, 30],
    'epochs': [50, 100],
    'model__optimizer': ['adam', 'sgd'], # stochastic
    'model__loss': ['binary_crossentropy', 'hinge'],
    'model__kernel_initializer': ['random_uniform', 'normal'],
    'model__activation': ['relu', 'tanh'], # tangente hiperbolica
    'model__neurons': [16, 8]
}
```
