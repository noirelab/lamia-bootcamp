```python
import numpy as np
import matplotlib.pylab as plt
import matplotlib
```

```python
# dados para o plot
data = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * data)

fig, ax = plt.subplots()
ax.plot(data, s)

ax.set(xlabel = 'time(s)', ylabel='voltage(mV)', title = 'mais simples impossvel, galera')
ax.grid()
```
