```python
import random
import numpy as np
```

```python
# cria um conjunto de dados com N pontos e k clusters
def createClusteredData(N, k):
    random.seed(10)
    pointsPerCluster = float(N)/k
    X = []
    # cria k clusters
    for i in range(k):
        incomeCentroid = random.uniform(20000.0, 200000.0)
        ageCentroid = random.uniform(20.0, 70.0)
        for j in range(int(pointsPerCluster)):
            X.append([random.normalvariate(incomeCentroid, 10000.0), random.normalvariate(ageCentroid, 2.0)])

    X = np.array(X)
    return X
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale

date = createClusteredData(100, 5)
model = KMeans(n_clusters=5)

# scalando pra normalizaer
model = model.fit(scale(date))

print(model.labels_)

# visualizar
plt.figure(figsize=(8, 6))
plt.scatter(date[:,0], date[:,1], c=model.labels_.astype(np.float_))
plt.show()
```
