```python
import numpy as np
import pandas as pd
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns
```

# Matplotlib

```python
# gráfico padrão
x = np.arange(-5, 5, 0.1)

plt.plot(x, norm.pdf(x))
plt.show()
```

Plots múltiplos e manipulação de eixos

```python
axes = plt.axes() # cria um objeto de eixos
axes.grid() # ativa a grade

# define os limites dos eixos
axes.set_xlim([-5, 5]) # define os limites dos eixo x
axes.set_ylim([0, 1.0]) # define os limites dos eixo y
axes.set_xticks(np.arange(-5, 6, 1)) # define os valores dos marcadores do eixo x
axes.set_yticks(np.arange(0, 1.1, 0.1)) # define os valores dos marcadores do eixo y

# customização de cores e linhas
plt.plot(x, norm.pdf(x), 'b-') # linha azul sólida
plt.plot(x, norm.pdf(x, 1.0, 0.5), 'r:') # linha vermelha pontilhada

# adiciona uma legenda
plt.legend(['vermelho', 'azul'], loc=4) # loc define qual canto vai aparecer

# mostra o gráfico
plt.show()
```

Pie chart

```python
values = [7, 2, 1, 1, 1, 1] 
colors = ['b', 'm', 'r', 'c', 'g', 'y']
explode = [0, 0, 0.2, 0, 0, 0] # destaca a terceira fatia
labels = ['verstappen', 'hamilton', 'sainz', 'leclerc', 'norris', 'piastri']

plt.pie(values, colors=colors, labels=labels, explode=explode)
plt.title('F1 2024 Wins so far')
plt.show()
```

Bar chart

```python
plt.bar(range(0, 6), values, color=colors) # cria um gráfico de barras
plt.show()
```

Scatter Plot

```python
x = np.random.randn(250)
y = np.random.randn(250)

plt.scatter(x, y)
plt.show()
```

BoxPlot

```python
data = np.random.randint(0, 10000, 5000)
plt.boxplot(data)
plt.show()
```

# Seaborn

```python
%matplotlib inline
col = 'Hamilton Verstappen Vettel Bottas Sainz Piastri'.split()
wins = [105, 61, 53, 10, 5, 1]
age = [39, 26, 37, 34, 29, 23]

df = pd.DataFrame({'Driver': col, 'Wins': wins, 'Age': age}) # cria um DataFrame
df
```

```python
vitorias = df['Wins'] # seleciona a coluna de vitórias
vitorias.plot(kind='bar') # cria um gráfico de barras pelo sns, mais bonito
```

```python
sns.distplot(df['Age']) # cria um gráfico de distribuição
```

```python
sns.pairplot(df, height=2.5) # cria um gráfico de pares
```

```python
sns.scatterplot(x = 'Age', y = 'Wins', data = df) # cria um gráfico scatter
```

```python
sns.jointplot(x = 'Age', y = 'Wins', data = df) # o jointplot cria um scatterplot e um histograma
```

```python
sns.lmplot(x='Age', y='Wins', data=df) # cria um gráfico de regressão linear
```

Boxplots

```python
# o meu dataframe não é o melhor pra boxplot, mas ainda continua sendo um exemplo
sns.set(rc={'figure.figsize':(8, 5)}) # define o tamanho do gráfico

ax = sns.boxplot(x='Driver', y='Wins', data=df) # cria um gráfico de caixa
ax.set_xticklabels(ax.get_xticklabels(), rotation=45) # rotaciona os rótulos do eixo x
```

```python
# swarm plot
ax = sns.swarmplot(x='Driver', y='Wins', data=df) # cria um gráfico de enxame, se o dataframe fosse mais apropriado, seria como um enxame de dados em cada grupo/coluna
ax.set_xticklabels(ax.get_xticklabels(), rotation=45) # rotaciona os rótulos do
```

```python
# countplot
# de novo, não é o melhor dataframe, mas ainda é um exemplo
ax = sns.countplot(x='Wins', data=df) # cria um gráfico de contagem
ax.set_xticklabels(ax.get_xticklabels(), rotation=45) # rotaciona os rótulos do eixo x
```
