```python
from bs4 import BeautifulSoup
import requests
import time
```

## find_all():

```python
with open('index.html') as html_file:
    content = html_file.read()
    soup = BeautifulSoup(content, 'lxml') # content é o conteúdo do arquivo html e lxml é o parser
    
    # print(soup.prettify()) # printa o html formatado
    courses_html_tags = soup.find_all('h5') # retorna uma array com todas as tags h5
    
    for course in courses_html_tags:
        print(course.text) # printa o texto da tag h5
```

## grabbing prices

```python
with open('index.html') as html_file:
    content = html_file.read()
    soup = BeautifulSoup(content, 'lxml')
    
    # tem que ser class_ pq class já é uma função do python
    course_cards = soup.find_all('div', class_='card') # retorna uma array com todas as tags div com a classe card
    
    for card in course_cards:
        course_name = card.h5.text
        course_price = card.a.text.split()[-1] # peg ao último elemento do array do split, que é o do XX$
        
        print(f'{course_name} custa {course_price}')
```

## looping through similars

```python
html_text = requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&searchTextSrc=&searchTextText=&txtKeywords=python&txtLocation=').text
# print(html_text) # response 200 significa que deu certo
soup = BeautifulSoup(html_text, 'lxml')
jobs = soup.find_all('li', class_='clearfix job-bx wht-shd-bx')

for job in jobs: # pega todas as vagas de emprego
    date = job.find('span', class_='sim-posted').text
    if 'few' in date: # se tiver few na data, printa
        empresa = job.find('h3', class_='joblist-comp-name').text.replace(' ', '')
        skill = job.find('span', class_='srp-skills').text.replace('  ', '')
        
        print(f'''
        Nome da empresa: {empresa}
        Habilidades necessárias: {skill}
        ''')
        print('')
```

## Prettifying the jobs paragraph / Job filtrating

```python
html_text = requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&searchTextSrc=&searchTextText=&txtKeywords=python&txtLocation=').text
soup = BeautifulSoup(html_text, 'lxml')
jobs = soup.find_all('li', class_='clearfix job-bx wht-shd-bx')

del_skill = input('filtro a ser deletado: ')
print(f'Filtrando resultados sem {del_skill}...\n')

for job in jobs: # pega todas as vagas de emprego
    date = job.find('span', class_='sim-posted').text
    
    if 'few' in date: # se tiver few na data, printa
        empresa = job.find('h3', class_='joblist-comp-name').text.replace(' ', '')
        skill = job.find('span', class_='srp-skills').text.replace('  ', '')
        link = job.header.h2.a['href'] # pega o link da vaga acessando um por um até chegar no href
        
        if del_skill not in skill: # se o filtro n estiver nas habilidades, printa
            print(f'empresa: {empresa.strip()}') # strip tira os espaços, melhora significavelmente a visualização
            print(f'habilidades: {skill.strip()}')
            print(f'link: {link.strip()}')
            print('')
```

## Automatização / Salvar em outro arquivo

```python
def find_jobs(): # estava prestes a criar uma eu mesmo e fugir do tutorial
    html_text = requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&searchTextSrc=&searchTextText=&txtKeywords=python&txtLocation=').text
    soup = BeautifulSoup(html_text, 'lxml')
    jobs = soup.find_all('li', class_='clearfix job-bx wht-shd-bx')

    del_skill = input('filtro a ser deletado: ')
    print(f'Filtrando resultados sem {del_skill}...\n')

    for index, job in enumerate(jobs): # pega todas as vagas de emprego
        date = job.find('span', class_='sim-posted').text
        
        if 'few' in date: # se tiver few na data, printa
            empresa = job.find('h3', class_='joblist-comp-name').text.replace(' ', '')
            skill = job.find('span', class_='srp-skills').text.replace('  ', '')
            link = job.header.h2.a['href'] # pega o link da vaga acessando um por um até chegar no href
            
            if del_skill not in skill: # se o filtro n estiver nas habilidades, printa
                with open(f'texistes/{index}.txt', 'w') as f: # cria um arquivo de texto com o nome do index
                    f.write(f'empresa: {empresa.strip()}\n') 
                    f.write(f'habilidades: {skill.strip()}\n')
                    f.write(f'link: {link}\n')
                    
                print(f'arquivo {index} salvo com sucesso')  
```

```python
if __name__ == '__main__':
    while True:
        find_jobs() # se o script for executado diretamente, chama a função
        minutos = 10
        print(f'Esperando {minutos} minutos...')
        time.sleep(minutos * 60) # espera 10 minutos
```
